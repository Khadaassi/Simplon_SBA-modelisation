```python
df['MIS_Status'] = df['MIS_Status'].map({'P I F': 1, 'CHGOFF': 0})  # 1 = remboursÃ©, 0 = dÃ©faut
```

```python
# Modifier les annÃ©es en fonction de la condition
df['ApprovalDate'] = df['ApprovalDate'].map(
    lambda x: x.replace(x[-3:], '-19' + x[-2:]) if int(x[-2:]) >= 50 else x.replace(x[-3:], '-20' + x[-2:])
)
df['ApprovalDate'] = pd.to_datetime(df['ApprovalDate'])
```

```python
df['Year']=df['ApprovalDate'].dt.year
```

```python
# CrÃ©ation de la figure et des axes pour 2 sous-graphiques en colonne (2 lignes, 1 colonne)
fig, axes = plt.subplots(2, 1, figsize=(20, 10))

# Premier scatter plot
sns.scatterplot(
    x='ApprovalDate', 
    y='DisbursementGross', 
    hue='MIS_Status', 
    data=df_filtered, 
    palette='coolwarm', 
    ax=axes[0]  # Utilisation de axes[0] au lieu de axes[0, 0]
)
axes[0].set_xlabel('ApprovalDate')
axes[0].set_ylabel('DisbursementGross')
axes[0].set_title('Graphique 1')

# DeuxiÃ¨me scatter plot
sns.scatterplot(
    x='ApprovalDate', 
    y='DisbursementGross', 
    hue='MIS_Status', 
    data=df_filtered2, 
    palette='coolwarm',  
    ax=axes[1]  # Utilisation de axes[1] au lieu de axes[0, 1]
)
axes[1].set_xlabel('ApprovalDate')
axes[1].set_ylabel('DisbursementGross')
axes[1].set_title('Graphique 2')

plt.tight_layout()  # Ajuste l'affichage pour Ã©viter le chevauchement
plt.show()

```
